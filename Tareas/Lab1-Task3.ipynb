# Task 3 – Métricas de Desempeño

En este notebook implementaremos las métricas **RMSE** (Root Mean Squared Error) y **MAE** (Mean Absolute Error) desde cero.

## 1. Importar librerías necesarias

```python
import numpy as np
```

## 2. Definir los datos

```python
# Valores reales
y_real = [100, 150, 200, 250, 300]

# Predicciones (note el error masivo en el último dato)
y_pred = [110, 140, 210, 240, 500]

print("Valores reales:", y_real)
print("Predicciones:  ", y_pred)
```

## 3. Implementar RMSE (Root Mean Squared Error)

**Fórmula:**
$$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_{real_i} - y_{pred_i})^2}$$

```python
def calcular_rmse(y_real, y_pred):
    """
    Calcula el Root Mean Squared Error (RMSE)
    
    Parámetros:
    y_real: lista de valores reales
    y_pred: lista de predicciones
    
    Retorna:
    float: valor de RMSE
    """
    # Convertir a arrays de numpy para facilitar operaciones
    y_real = np.array(y_real)
    y_pred = np.array(y_pred)
    
    # Calcular la diferencia entre valores reales y predicciones
    diferencias = y_real - y_pred
    
    # Elevar al cuadrado cada diferencia
    diferencias_cuadrado = diferencias ** 2
    
    # Calcular la media de los cuadrados
    mse = np.mean(diferencias_cuadrado)
    
    # Obtener la raíz cuadrada
    rmse = np.sqrt(mse)
    
    return rmse
```

## 4. Implementar MAE (Mean Absolute Error)

**Fórmula:**
$$MAE = \frac{1}{n}\sum_{i=1}^{n}|y_{real_i} - y_{pred_i}|$$

```python
def calcular_mae(y_real, y_pred):
    """
    Calcula el Mean Absolute Error (MAE)
    
    Parámetros:
    y_real: lista de valores reales
    y_pred: lista de predicciones
    
    Retorna:
    float: valor de MAE
    """
    # Convertir a arrays de numpy para facilitar operaciones
    y_real = np.array(y_real)
    y_pred = np.array(y_pred)
    
    # Calcular la diferencia entre valores reales y predicciones
    diferencias = y_real - y_pred
    
    # Obtener el valor absoluto de cada diferencia
    diferencias_absolutas = np.abs(diferencias)
    
    # Calcular la media de los valores absolutos
    mae = np.mean(diferencias_absolutas)
    
    return mae
```

## 5. Calcular y comparar ambas métricas

```python
# Calcular RMSE
rmse = calcular_rmse(y_real, y_pred)
print(f"RMSE: {rmse:.2f}")

# Calcular MAE
mae = calcular_mae(y_real, y_pred)
print(f"MAE:  {mae:.2f}")

print("\n" + "="*60)
```

## 6. Análisis de errores individuales

```python
# Analizar los errores individuales
print("Análisis de errores individuales:")
print("-" * 60)
for i in range(len(y_real)):
    error = y_real[i] - y_pred[i]
    error_abs = abs(error)
    error_cuadrado = error ** 2
    print(f"Dato {i+1}: Real={y_real[i]}, Pred={y_pred[i]}, Error={error}, |Error|={error_abs}, Error²={error_cuadrado}")
```

## 7. Conclusión: Comparación de métricas

```python
print("\n" + "="*60)
print("CONCLUSIÓN:")
print("="*60)
print(f"\nRMSE: {rmse:.2f}")
print(f"MAE:  {mae:.2f}")
print(f"\nDiferencia (RMSE - MAE): {rmse - mae:.2f}")
print("\n¿Cuál métrica penalizó más el error del último dato (500)?")
print("-" * 60)
print(f"El RMSE ({rmse:.2f}) penalizó MÁS que el MAE ({mae:.2f}) el error del último dato.")
print("\n¿Por qué esto es importante?")
print("-" * 60)
print("""
Razones:
1. El RMSE eleva al CUADRADO los errores, lo que hace que errores grandes 
   (como el 200 del último dato) tengan un impacto DESPROPORCIONADAMENTE mayor.
   
2. El error del último dato (300 - 500 = -200) se convierte en:
   - MAE: |−200| = 200 (penaliza linealmente)
   - RMSE: (−200)² = 40,000 (penaliza cuadráticamente)

3. En el contexto de DOSIS DE MEDICAMENTOS:
   - Un error grande puede ser FATAL (sobredosis o subdosis crítica)
   - El RMSE es MÁS APROPIADO porque penaliza severamente los errores grandes
   - Queremos un modelo que evite errores extremos, incluso si tiene más 
     errores pequeños
   - Una dosis incorrecta por 200 unidades podría matar al paciente
   
4. Conclusión para predicción médica:
   - RMSE es preferible cuando errores grandes son INACEPTABLES
   - MAE puede ser engañoso al no reflejar la gravedad de errores extremos
   - En medicina, seguridad > precisión promedio
""")
```
