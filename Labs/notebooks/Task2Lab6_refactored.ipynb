# Task 2 - Connect Four AI (Refactored)

This notebook imports the reusable baseline package from `src/connect4` and keeps the same visible demos from the original lab notebook.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC = ROOT / 'src'
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from connect4 import (
    AlphaBetaAgent,
    Connect4,
    MinimaxAgent,
    SmartAlphaBetaAgent,
    display_board,
    evaluate,
    play_ai_vs_random,
    play_human_vs_ai,
    print_board,
    run_ai_vs_random_series,
)
```

## Demo: empty board

```python
game = Connect4()
display_board(game, title='Tablero vacio')
print(f'Movimientos validos: {game.actions()}')
print(f'Es terminal?: {game.is_terminal()}')
```

## Demo: Minimax and Alpha-Beta

```python
test_game = Connect4()
test_game.drop_piece(3)
test_game.drop_piece(2)
test_game.drop_piece(3)
test_game.drop_piece(4)
test_game.drop_piece(5)

print_board(test_game)
print(f"Turno de: {'X (P1)' if test_game.current_player == 1 else 'O (P2)'}")
print('=' * 50)

minimax_agent = MinimaxAgent(ai_player=Connect4.PLAYER2, depth=4)
col_mm, nodes_mm = minimax_agent.get_best_move(test_game.copy())
print('Minimax puro (d=4):')
print(f'  Mejor columna: {col_mm}')
print(f'  Nodos visitados: {nodes_mm:,}')

ab_agent = AlphaBetaAgent(ai_player=Connect4.PLAYER2, depth=4)
col_ab, nodes_ab = ab_agent.get_best_move(test_game.copy())
print('Alfa-Beta (d=4):')
print(f'  Mejor columna: {col_ab}')
print(f'  Nodos visitados: {nodes_ab:,}')
```

## Demo: heuristic evaluation

```python
test_game2 = Connect4()
test_game2.drop_piece(3)
test_game2.drop_piece(0)
display_board(test_game2, title='Prueba heuristica')
print(f'Evaluacion para P2 (O): {evaluate(test_game2, Connect4.PLAYER2)}')
print(f'Evaluacion para P1 (X): {evaluate(test_game2, Connect4.PLAYER1)}')
```
